Podemos ver que los parasitos que se detectan son 24 mientras que solo detecta 6 nucleos. Para el ejercicio de asignar un parasito a cada nucleo vamos 
a utilizar el metodo NEAREST NEIGHBOUR. Para esto utilizaremos librerias como pandas y numpy para manipular los datos y cKDTree para el metodo de clasificacion.

```python
import numpy as np
import pandas as pd
from scipy.spatial import cKDTree
### Cargamos los datos 
nu_df = pd.read_csv("coordsNU.csv")
para_df = pd.read_csv("coorsPARA.csv")
```

```python
nucleos = nu_df[["X","Y"]].to_numpy()
parasitos = para_df[["X","Y"]].to_numpy()

# Construir el KDTree
tree = cKDTree(nucleos)
dists, idxs = tree.query(parasitos, k=1)

# Convertir parasitos a DataFrame
resultado = pd.DataFrame(parasitos, columns=["X", "Y"])
resultado = resultado.reset_index().rename(columns={"index": "parasite_id"})

# Agregar columnas con los resultados
resultado["nearest_nucleus_id"] = idxs
resultado["distancia"] = dists
print(resultado)
```
